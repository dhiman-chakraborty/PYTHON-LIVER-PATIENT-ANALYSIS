```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r'D:\PYTHON PROJECTS\LIVER PATIENT ANALYSIS\DATASET\liver_patient_dataset.csv',encoding='latin1')

print(df.head(10))
```

```python

# Standardize column names
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')

# Rename target column
df = df.rename(columns={'selector': 'target'})

# Convert target to binary
df['target'] = df['target'].map({
    'Liver Disease': 1,
    'No Liver Disease': 0
})

# Encode gender
df['gender'] = df['gender'].map({'Male': 1, 'Female': 0})

# Check missing values
print(df.isnull().sum())
```

```python
# Fill missing values (if any)
df = df.fillna(df.median(numeric_only=True))

# Remove duplicates
df = df.drop_duplicates()

print("Cleaned Shape:", df.shape)
print(df.head())
```

```python
### Feature Engineering ###
# Create ratio features
df['sgpt_sgot_ratio'] = df['sgpt'] / (df['sgot'] + 1)

# Age group segmentation
df['age_group'] = pd.cut(df['age'],bins=[0, 30, 50, 70, 100],labels=['Young', 'Middle', 'Senior', 'Old'])

print(df.head(20))
```
